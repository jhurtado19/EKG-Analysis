```python
# pass in array of values and limit
# will return the average of the peak values above the limit

def averageGreater(array, lim):
  sum = 0
  length = 0
  i = 1
  plt.plot(array)
  print(f"Peak values greater than {lim}")
  while i < len(array) - 1:
    if array[i] > lim and array[i] >= array[i-1] and array[i] >= array[i+1]:
      print(array[i])
      plt.vlines(i,array[i],-array[i])
      sum += array[i]
      length += 1
      while array[i] == array[i+1]:
        i += 1
    i += 1
  print(f"Average: {sum / length}")
```

```python
averageGreater(data, 7000)
```
